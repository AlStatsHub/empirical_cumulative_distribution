```python
import numpy as np
from matplotlib import pyplot as plt

from scipy.stats import gamma, norm
np.random.seed(42) # set seed value to 42 for reproducibility

# Generate three different random variables
x1 = gamma.rvs(10, scale=2, size=1000)
x2 = gamma.rvs(20, scale=4, size=1000)
x3 = norm.rvs(loc=0, scale=5, size=1000)

# Combine the three variables into a single sample
sample = np.hstack((x1, x2, x3))

# Plot the histogram of the sample
plt.hist(sample, bins=50)
plt.show()

# Fit the empirical cumulative distribution function (ECDF) to the sample
from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(sample)

# Plot the ECDF of the sample
plt.plot(ecdf.x, ecdf.y)
plt.show()
# Calculate the probabilities of various values using the ECDF
# Calculate the probabilities of various values using the ECDF
probs = [-10, -5, 0, 5, 10, 15, 20, 25, 30, 35]
for p in probs:
    print('{} {:.4f}'.format(p, ecdf(p)))
```
